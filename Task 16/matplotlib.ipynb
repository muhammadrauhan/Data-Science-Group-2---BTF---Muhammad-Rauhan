## Learn about Basic plots (Line, Scatter & Bar):

Time to visually observe and find insights, so we will be learning Matplotlib. It is particularly useful for generating graphs and charts to visualize data. Matplotlib is highly customizable, offering a variety of plot types including line plots, scatter plots, bar charts, histograms, and more.

### Import Library:

```python
import matplotlib.pyplot as plt
```

### 1. Basic Plots:


**Line Plot** is one of the easiest and most basic graphical analysis techniques that play an important role in data analysis when working on machine learning or data science projects. They are used to express a relationship between two variables.

```python
# define a data

sales_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales_A': [200, 220, 210, 230, 250, 270, 290, 310, 300, 320, 330, 340],
    'Sales_B': [180, 190, 200, 220, 240, 250, 260, 280, 290, 300, 310, 320],
    'Sales_C': [150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 260]
}
```

##### Create simple line plot:

```python
plt.figure(figsize=(10, 6))
plt.plot(sales_data['Month'], sales_data['Sales_A'], label = 'Sales A',
         color = 'purple')
plt.show()
```

##### Add marker to a line plot:

```python
plt.figure(figsize=(10, 6))
plt.plot(sales_data['Month'], sales_data['Sales_A'], label = 'Sales A',
         color = 'purple', marker = 'o')
plt.show()
```

##### Plotting multiple lines:

```python
# create a plot fot Sales_A, Sales_B & Sales_C

plt.figure(figsize=(10, 6))

plt.plot(sales_data['Month'], sales_data['Sales_A'], label='Sales A', color='purple', linestyle='-', marker='o')
plt.plot(sales_data['Month'], sales_data['Sales_B'], label='Sales B', color='cyan', linestyle='--', marker='s')
plt.plot(sales_data['Month'], sales_data['Sales_C'], label='Sales C', color='red', linestyle='-.', marker='^')
plt.show()
```

#####  Customize plot:

```python
plt.figure(figsize=(10, 6))
plt.plot(sales_data['Month'], sales_data['Sales_A'], label='Sales A', color='purple', linestyle='-', marker='o')
plt.plot(sales_data['Month'], sales_data['Sales_B'], label='Sales B', color='cyan', linestyle='--', marker='s')
plt.plot(sales_data['Month'], sales_data['Sales_C'], label='Sales C', color='red', linestyle='-.', marker='^')

plt.title('Monthly Sales Data (Multiple Products)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend() # on top left corner
plt.grid(True)
plt.show()
```

### 2. Scatter Plots:

A **scatter plot** is a diagram where each value in the data set is represented by a dot.

```python
# define a data
data = {
    'Age': [23, 45, 56, 34, 25, 67, 78, 34, 45, 23, 55, 60],
    'Height': [165, 170, 175, 180, 155, 160, 185, 190, 175, 160, 170, 175],
    'Weight': [65, 70, 75, 80, 55, 60, 85, 90, 75, 60, 70, 75]
}
```

##### Create basic scatter plot:

```python
# create a plot for Age vs. Height
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Height'])
plt.show()
```

##### Customize marker styles, sizes, and colors:

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Height'], 
            color='purple', 
            marker='o', 
            s=100) # s is marker size
plt.show()
```

##### Plot multiple scatter plots on the same graph &
##### Add labels and titles to the scatter plot:

```python
# x and y values for first and second scatter plot
x1 = data['Age']
y1 = data['Height']

x2 = data['Age']
y2 = data['Weight']

# first scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x1, y1, color='purple', marker='o', 
            s=100)

#second scatter plot
plt.scatter(x2, y2, color='skyblue', marker='d', 
            s=100)

# customize plot
plt.title('Multiple Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Weight (kg) and Height (cm)')
plt.grid(True)
plt.show()
```

### 3. Barplots:

**Bar plots** are a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point, and the width of each bar represents the category of the data.

```python
# define a sales data
sales_2 = {
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Food', 'Toys'],
    'Revenue': [50000, 40000, 30000, 20000, 10000],
    'Profit': [12000, 8000, 6000, 4000, 2000]
}
```

##### Creating a vertical bar chart:

```python
plt.figure(figsize=(10, 6))
plt.bar(sales_2['Category'], sales_2['Revenue'])
plt.xlabel('Category')
plt.ylabel('Revenue ($)')
plt.show()
```

##### Creating a horizontal bar chart:

```python
plt.figure(figsize=(10, 6))
plt.barh(sales_2['Category'], sales_2['Profit'])
plt.xlabel('Profit ($)')
plt.ylabel('Category')
plt.show()
```

##### After Customization on both plots:

```python
# first bar chart
plt.figure(figsize=(10, 6))
plt.bar(sales_2['Category'], sales_2['Revenue'], color='orange', 
        edgecolor='red')

plt.title('Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Revenue ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# second bar chart
plt.figure(figsize=(10, 6))
plt.barh(sales_2['Category'], sales_2['Profit'], color='skyblue', 
         edgecolor='magenta')

plt.title('Profit by Category')
plt.xlabel('Profit ($)')
plt.ylabel('Category')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

### 4. Pie Charts:

**Pie charts** show the size of items (called wedge) in one data series, proportional to the sum of the items. The data points in a pie chart are shown as a percentage of the whole pie.

```python
# define a data
data2 = {
    'Categories': ['Electronics', 'Furniture', 'Clothing', 'Food', 'Toys'],
    'Values': [45000, 30000, 20000, 15000, 10000]
}
```

##### Basic pie chart:

```python
plt.figure(figsize=(8, 8))
plt.pie(data2['Values'], labels = data2['Categories'])
plt.title('Category Distribution')
plt.show()
```

##### Cutomize slice colors and add a legend:

```python
colors = ['skyblue', 'orange', 'lightgreen', 'lightcoral', 'yellow']
plt.figure(figsize=(8, 8))
plt.pie(data2['Values'], labels = data2['Categories'], 
        colors=colors)

# Add a legend
plt.legend(data2['Categories'], title="Categories", loc="best")
plt.title('Category Distribution')
plt.show()
```

**loc="best"**: This parameter determines the location of the legend. The "best" option places the legend in the position that the matplotlib library determines to be the least obstructive or most aesthetically pleasing relative to the data in the plot.

##### Explode a slice of the pie chart &
##### Adding percentage labels to the slices:

```python
# explode the first slice (Electronics)
explode = (0.1, 0, 0, 0, 0)

plt.figure(figsize=(8, 8))
plt.pie(data2['Values'], labels = data2['Categories'], 
        colors=colors, 
        explode=explode,
        autopct='%1.1f%%') # add percentage labels

plt.legend(data2['Categories'], title="Categories", loc="best")
plt.title('Category Distribution (Exploded Electronics)')
plt.show()
```
