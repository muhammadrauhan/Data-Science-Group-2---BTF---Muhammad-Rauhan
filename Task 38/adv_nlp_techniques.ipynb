* ## Named Entity Recognition (NER):

**Named Entity Recognition (NER)** is basically where you try to extract different entities in yout text. <br>
Let's explore NER support in spaCy.

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
nlp.pipe_names
```

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")

for ent in doc.ents:
    print(ent.text, " | ", ent.label_, " | ", spacy.explain(ent.label_))
```

```python
from spacy import displacy

displacy.render(doc, style="ent")
```

Twitter it does not recognize it because SpaCy using some kind of rules or patterns, to recognize companies. <br>

Let's say 'Twitter Inc' that's a popular way of representing companies in U.S.

```python
doc = nlp("Tesla Inc is going to acquire Twitter Inc for $45 billion")

for ent in doc.ents:
    print(ent.text, " | ", ent.label_, " | ", spacy.explain(ent.label_))
```

```python
displacy.render(doc, style="ent")
```

Realizing that SpaCy out of the box **Named Entity Recognition (NER)** is not perfect. And that's perfectly fine. It is using some kind of rules that if there is a word with first letter as capital, followed by Inc then that's a company.
