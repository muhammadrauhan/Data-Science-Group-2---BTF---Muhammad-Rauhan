# Task 19: Statistical Plotting with Seaborn

Seaborn is generally better suited for creating statistical graphics and exploratory data analysis (EDA). Time to visualize (regression plots, distribution plots etc) and level up plotting techniques which inlcudes **Statistical Plots**.
Some statistical plots are;
- Regression plot
- Residual plot
- Kernel Density Estimate plot
- Distribution plot
- Violin plot
- and more....

## Datasets:

- **iris.csv** -- The Iris Dataset consists of 150 records of iris flowers, each with four features: sepal length, sepal width, petal length, and petal width, measured in centimetres.

- **tips.csv** -- The Tips Dataset contains information about the tips given in restaurants.

## Importing Libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Dataset 1:

```python
iris = pd.read_csv("iris.csv")
iris
```

### 1. Histogram:

- Histograms visualize the distribution of a continuous variable.
- In a histogram, the data is divided into bins and the height of each bin represents the frequency or count of data points within that bin.
- In Seaborn, histograms can be created using the histplot() function.

```python
# basic histogram
plt.figure(figsize=(10, 6))
sns.histplot(x='petal_length', data=iris)
```

```python
# customized histogram

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
ax = sns.histplot(data=iris, x="petal_length", bins=20, kde=True, color="green")
ax.set(xlabel="Petal Length (cm)", ylabel="Frequency")
ax.set_title("Distribution of Petal Lengths in Iris Flowers")
```

### 2. Violin Plot:

- A violin plot is a type of data visualization that combines aspects of both box plots and density plots.
- The width of the violin represents the density estimate, with wider parts indicating higher density, and the IQR and median are shown as a white dot and line within the violin.
- In Seaborn, violin plot can be created using the violinplot() function.

```python
plt.figure(figsize=(10, 6))
sns.set_style("darkgrid")
ax1 = sns.violinplot(x="species", y="petal_length", data=iris)
ax1.set(ylabel="Petal Length (cm)")
ax1.set_title("Petal Length by Species")
```

### 3. Pair Plot:

- Pair plots are a type of visualization in which multiple pairwise scatter plots are displayed in a matrix format.
- Each scatter plot shows the relationship between two variables, while the diagonal plots show the distribution of the individual variables.
- In Seaborn, pair plot can be created using the pairplot() function.

```python
# simple pair plot 
sns.pairplot(data=iris)
```

```python
# customized pair plot
sns.pairplot(data=iris, hue="species", diag_kind="kde", palette="husl")
plt.title('Iris Dataset Pair Plot')
```

- **hue** parameter specifies the column to use for color encoding.
- **diag_kind** parameter specifies the type of plot to use for the diagonal subplots.
- **palette** parameter specifies the color palette to use.

### 4. Joint Plot:

- Joint plot is a powerful visualization technique in seaborn that combines two different plots in one visualization: a scatter plot and a histogram.
- The scatter plot shows the relationship between two variables, while the histogram shows the distribution of each individual variable.
- In Seaborn, joint plot can be created using the jointplot() function.

```python
sns.set_style('whitegrid')
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
```

### 5. Residual Plot:

- Plot the residuals of a linear regression
- This function will regress y on x (possibly as a robust or polynomial regression) and then draw a scatterplot of the residuals.

```python
plt.figure(figsize=(10, 6))
ax2 = sns.residplot(data=iris, x='sepal_length', y='petal_length')
ax2.set_xlabel('Sepal Length')
ax2.set_ylabel('Residuals of Petal Length')
ax2.set_title('Residual Plot of Petal Length vs Sepal Length')
```

### 6. Distribution Plot:

```python
plt.figure(figsize=(10, 6))
ax_1 = sns.distplot(iris['sepal_length'], kde=True, bins=20)
ax_1.set_xlabel('Sepal Length')
ax_1.set_ylabel('Density')
ax_1.set_title('Distribution Plot of Sepal Length')
```

## Loading the Dataset 2:

```python
tips = pd.read_csv("E:\Data Science BTF\Task 18/tips.csv")
tips
```

### 7. Facet Grid:

- FacetGrid is a powerful seaborn tool that allows you to visualize the distribution of one variable as well as the relationship between two variables.
- FacetGrid creates a grid of subplots based on the unique values in the categorical variable specified.

```python
g = sns.FacetGrid(tips, col="day")

# plot histogram for total_bill in each day
g.map(sns.histplot, "total_bill")
```

- A FacetGrid object is created using the "FacetGrid" function from Seaborn.
-  Finally, the "map" function is called on the FacetGrid object "g" to plot a histogram for the "total_bill" column in each day.

### 8. Kernel Density Estimate (KDE) Plot:

- Density plots, also known as kernel density plots, are a type of data visualization that display the distribution of a continuous variable.
- In Seaborn, density plots can be created using the kdeplot() function.

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=tips, x="total_bill")
```

```python
# customized kde plot

plt.figure(figsize=(10, 6))
ax3 = sns.kdeplot(data=tips, x="total_bill", hue="time", fill=True, alpha=0.6, linewidth=1.5)
ax3.set(xlabel="Total Bill ($)", ylabel="Density")
ax3.set_title("Density Plot of Total Bill by Meal Time")
```

- **"hue"** parameter to differentiate between "lunch" and "dinner" meal times.

- The **"fill"** parameter to fill the area under the curve.

- The **"alpha"** and **"linewidth"** parameters to make the plot more visually appealing.

### 9. Heatmap:

- A heatmap is a graphical representation of data that uses colors to depict the value of a variable in a two-dimensional - space.
- Heatmaps are commonly used to visualize the correlation between different variables in a dataset.

```python
corr = tips.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr)
```

- The resulting plot shows the correlation between different variables in the "tips" dataset.
- The darker the color, the stronger the correlation between the variables.

### 10. Linear Regression Plot:

```python
plt.figure(figsize=(10, 6))
ax4 = sns.regplot(data=tips, x='total_bill', y='tip')
ax4.set_xlabel('Total Bill')
ax4.set_ylabel('Tip')
ax4.set_title('Linear Regression of Tip vs Total Bill')
```

### 11.  Multiple Linear Regression Plot:

```python
plt.figure(figsize=(10, 6))
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```
