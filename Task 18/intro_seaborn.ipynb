# Task 18: Introduction to Seaborn (Visualization with Seaborn)

**What is Seaborn?** Built on top of Matplotlib, Seaborn is a well-known Python library for data visualization that offers an easy-to-use interface for creating informative and aesthetically pleasing statistical graphics/plots.

## Seaborn Plot Types:

Seaborn provides a wide range of plot types that can be used for data visualization and exploratory data analysis (EDA). Any visualization of Seaborn can fall into one of these three categories,

- **Univariate** – x only (contains only one axis of information).
- **Bivariate** – x and y (contains two axis of information).
- **Trivariate** – x, y, z (contains three axis of information).

## Datasets:

- **tips.csv** -- It contains information about the tips given in restaurants.

- **titanic.csv** -- It contains information of all the passengers aboard the RMS Titanic, which unfortunately was shipwrecked.

### Importing Libraries:

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading the Dataset 1:

```python
tips = pd.read_csv("tips.csv")
tips
```

### 1. Scatter Plot:

- Scatter plots are used to visualize the relationship between two continuous variables.
- In Seaborn, scatter plots can be created using the scatterplot() function.

```python
# basic scatter plot

plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
# customized scatter plot

plt.figure(figsize=(10, 6))
ax = sns.scatterplot(x='total_bill', y='tip', hue='sex', size='size', sizes = (50 ,200), data=tips)
ax.set_title('Relationship Between Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

**You can also use plt.xlabel() and plt.ylabel() from matplotlib for labelling.**
- **'hue=sex'** is used color code the points.
- **'size=size'** is used to adjust the size of the points.
- **'sizes parameter'** specifying the range of sizes to use.

```python
# scatter plot with color palette

plt.figure(figsize=(10, 6))
sns.set_style('whitegrid')
ax1 = sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips, palette="Set2")
ax1.set_title('Total Bill vs Tip')
ax1.set_xlabel('Total Bill ($)')
ax1.set_ylabel('Tip ($)')
```

The **'palette parameter'** is set to "Set2" to specify the color scheme of the plot.

### 2. Box Plot:

- Box plots are a type of visualization that shows the distribution of a dataset.
- In Seaborn, box plots can be created using the boxplot() function.

```python
# box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# customized box plot

plt.figure(figsize=(10, 6))
sns.set_style('darkgrid')
ax2 = sns.boxplot(x='day', y='total_bill', hue='time', linewidth=1.5, fliersize=4, data=tips, palette='Set3')
ax2.set_title('Box Plot of Total Bill by Day and Meal Time')
ax2.set_xlabel('Day of the Week')
ax2.set_ylabel('Total Bill ($)')
```

The **'linewidth'** and **'fliersize'** parameters are used to adjust the appearance of the plot.

### 3. Line Plot:

- Line plots are used to visualize trends in data over time or other continuous variables.
- In Seaborn, line plots can be created using the lineplot() function.

```python
plt.figure(figsize=(10, 6))
sns.set_style('whitegrid')
sns.lineplot(x='total_bill', y='tip', hue='sex', markers=True, data=tips)
```

## Loading the Dataset 2:

```python
titanic = pd.read_csv('titanic.csv')
titanic
```

### 4. Bar Plot:

- Bar plots are used to visualize the relationship between a categorical variable and a continuous variable.
- In a bar plot, each bar represents the mean or median (or any aggregation) of the continuous variable for each category.
- In Seaborn, bar plots can be created using the barplot() function.

```python
# basic bar plot

plt.figure(figsize=(10, 6))
sns.barplot(x="class", y="fare", data=titanic)
```

```python
# customize the bar plot

plt.figure(figsize=(10, 6))
sns.set_style('whitegrid')
ax3 = sns.barplot(x="class", y="fare", hue="sex", ci=None, palette="muted", data=titanic)
ax3.set_title("Average Fare By Class and Gender on the Titanic")
ax3.set_xlabel("Class")
ax3.set_ylabel("Fare")
```

### 5. Count Plot:

- Countplot method is used to show the counts of observations in each categorical bin using bars.
- In Seaborn, count plot can be created by using countplot() function.

```python
plt.figure(figsize=(10, 6))
sns.set_style('whitegrid')
ax4 = sns.countplot(data=titanic, x='sex', palette='Set2')
ax4.set(xlabel='Sex', ylabel='Count')
ax4.set_title('Count of Passengers by Sex')
```
