```python
import gradio as gr
from transformers import pipeline
from newspaper import Article
from langdetect import detect
import torch

# Check if GPU is available and set the device accordingly
device = 0 if torch.cuda.is_available() else -1  # 0 for GPU, -1 for CPU

# Load the translation and summarization models with device argument
arabic_translator = pipeline("translation", model="Helsinki-NLP/opus-mt-ar-en", device=device)
chinese_translator = pipeline('translation',model="Helsinki-NLP/opus-mt-zh-en",device = device)
japanese_translator = pipeline('translation',model = "Helsinki-NLP/opus-mt-ja-en", device = device)
# urdu_translator = pipeline('translation',model = "Helsinki-NLP/opus-mt-en-ur", device = device)

summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=device)

```

```python
def chunk_text(text, max_length=400):  # Increased chunk size
    # Split text into chunks that fit the model's max length
    chunks = []
    for i in range(0, len(text), max_length):
        chunks.append(text[i:i + max_length])
    return chunks

# Function to fetch article and detect language
def fetch_and_process_article(url):
    article = Article(url)
    article.download()
    article.parse()

    original_news = article.text

    detected_language = detect(original_news)
    print(f"Detected Language: {detected_language}")

    if detected_language == "ar":  # Arabic
        translator = arabic_translator
    elif detected_language == "zh-cn" or detected_language == "zh-tw":  # Chinese
        translator = chinese_translator
    elif detected_language == "ja":  # Japanese
        translator = japanese_translator
    else:
        return "Language not supported", "", ""

    # Chunk the original news
    chunks = chunk_text(original_news)

    # Translate and summarize each chunk
    translated_chunks = []
    summaries = []

    for chunk in chunks:
        translated_text = translator(chunk, max_length=400, num_beams=4, early_stopping=True)[0]['translation_text']

        # Adjusting summarization lengths
        summary = summarizer(translated_text, max_length=30, min_length=10, do_sample=False)[0]['summary_text']

        translated_chunks.append(translated_text)
        summaries.append(summary)

    # Join the results
    translated_text = ' '.join(translated_chunks)
    summary = ' '.join(summaries)

    return original_news, translated_text, summary
```

```python
# Create Gradio interface
interface = gr.Interface(
    fn=fetch_and_process_article,
    inputs=gr.Textbox(label = 'Article URL'),
    outputs=[gr.Textbox(label= "Original Article"),gr.Textbox(label= "Translated Article"),gr.Textbox(label= "Summarized Article")],
    title="Translation and Summarization with Language Detection",
    description="Fetch an article, detect its language, translate to English, and summarize it."
)
```

```python
interface.launch(debug = True)
```
